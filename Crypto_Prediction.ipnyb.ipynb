```python
# create historical data
def create_dataset(dataset, look_back=3):
    dataX, dataY = [], []
    for i in range(len(dataset)-look_back-1):
        a = dataset[i:(i+look_back)]
        dataX.append(a)
        dataY.append(dataset[i + look_back])
    return np.array(dataX), np.array(dataY)
```

```python
# BTC Close
x_BTC, y_BTC = create_dataset(df_BTC_close['scale'])
print (x_BTC.shape)
print (y_BTC.shape)
```

```python
# data training and testing for BTC
size = int(len(x_BTC) * 0.70)

x_train_BTC, x_test_BTC = x_BTC[0:size], x_BTC[size:len(x_BTC)]
y_train_BTC, y_test_BTC = y_BTC[0:size], y_BTC[size:len(x_BTC)]


print ('x_train BTC',x_train_BTC.shape, x_train_BTC[:5],y_train_BTC[:5])
print ('x_test BTC',x_test_BTC.shape, x_test_BTC[:5],y_test_BTC[:5])

plt.plot(y_train_BTC)
plt.title('training forecasting target BTC')
```

```python
# perform mlp and evaluate using MSE
from sklearn.metrics import mean_squared_error
from sklearn.neural_network import MLPRegressor
clf = MLPRegressor(activation = 'tanh',solver='adam')
```

```python
# BTC
clf.fit(x_train_BTC,y_train_BTC)

train_mse_BTC = clf.predict(x_train_BTC)
test_mse_BTC = clf.predict(x_test_BTC)
print ('MSE training BTC', mean_squared_error(train_mse_BTC,y_train_BTC))
print ('MSE testing BTC', mean_squared_error(test_mse_BTC,y_test_BTC))
```

```python
# ETH
clf.fit(x_train_ETH,y_train_ETH)

train_mse_ETH = clf.predict(x_train_ETH)
test_mse_ETH = clf.predict(x_test_ETH)
print ('MSE training ETH', mean_squared_error(train_mse_ETH,y_train_ETH))
print ('MSE testing ETH', mean_squared_error(test_mse_ETH,y_test_ETH))
```

```python
# ADA
clf.fit(x_train_ADA,y_train_ADA)

train_mse_ADA = clf.predict(x_train_ADA)
test_mse_ADA = clf.predict(x_test_ADA)
print ('MSE training ADA', mean_squared_error(train_mse_ADA,y_train_ADA))
print ('MSE testing ADA', mean_squared_error(test_mse_ADA,y_test_ADA))
```

```python
# LINK
clf.fit(x_train_LINK,y_train_LINK)

train_mse_LINK = clf.predict(x_train_LINK)
test_mse_LINK = clf.predict(x_test_LINK)
print ('MSE training LIJK', mean_squared_error(train_mse_LINK,y_train_LINK))
print ('MSE testing LINK', mean_squared_error(test_mse_LINK,y_test_LINK))
```

```python
# DOT
clf.fit(x_train_DOT,y_train_DOT)

train_mse_DOT = clf.predict(x_train_DOT)
test_mse_DOT = clf.predict(x_test_DOT)
print ('MSE training DOT', mean_squared_error(train_mse_DOT,y_train_DOT))
print ('MSE testing DOT', mean_squared_error(test_mse_DOT,y_test_DOT))
```
