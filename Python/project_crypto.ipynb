```python
import pandas as pd
import glob

path = r'/Users/sowmyavemulapalli/Desktop/Data_Analyst/Assignments/Project/Cryptocurrencies/Resources'
# use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

data_df = pd.concat(li, axis=0, ignore_index=True)
```

```python
data_df.head()
```

```python
data_df["Difference"] = data_df["High"] - data_df["Low"]
```

```python
data_df.head()
```

```python
data_df["Volatility_percent"] = (data_df["Difference"]/data_df["Close"])*100
```

```python
data_df.loc[data_df['Volatility_percent'] <= 30, 'Volatility'] = 0 
data_df.loc[(data_df['Volatility_percent'] > 30) & (data_df['Volatility_percent'] <= 60), 'Volatility'] = 1
data_df.loc[data_df['Volatility_percent'] > 60 , 'Volatility'] = 2
```

```python
data_df["Daily_Difference"] = ((data_df["Close"]- data_df["Open"])/data_df["Open"] *100).round(2)
data_df.head()
```

```python
data_df.count()
```

```python
data_df.to_csv('cryptocurrencies.csv', index=False)
```
